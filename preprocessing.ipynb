## Replace NaNs with average for numerical columns

```python
def fill_average(df, col, mini=None, maxi=None):
    if mini is None and maxi is None:
        normal = [elem for elem in df[col] if not np.isnan(elem)]
        avg = sum(normal)/len(normal)
        def fill(elem):
            if np.isnan(elem):
                return avg
            else:
                return elem
        df[col] = [fill(elem) for elem in df[col]]        
    else:
        normal = [elem for elem in df[col] if not np.isnan(elem) and elem >= mini and elem <= maxi]
        avg = sum(normal)/len(normal)
        def fill(elem):
            if np.isnan(elem) or elem < mini or elem > maxi:
                return avg
            else:
                return elem
        df[col] = [fill(elem) for elem in df[col]]
    return df
```

## Removing columns with low correlation to happiness ladder

```python
df = del_cols(df, "year")
correlations = df.corr()['happiness_ladder'].drop('happiness_ladder').drop(addcols).to_dict()
print(len(correlations))
```

```python
thresholds = {
    'all': 0.0,
    'complex': 0.05,
    'medium': 0.1,
    'basic': 0.15,
    'simple': 0.25
}
```

```python
STATE = 42
cols_dict = dict()

for complexity, threshold in thresholds.items():
    filtered_cols = []
    for col, corr in correlations.items():
        if abs(corr) >= threshold:
            filtered_cols.append(col)
            
    cols = filtered_cols + addcols
    cols.append('happiness_ladder')
    print(f"{complexity}: {len(cols)}")
    cols_dict[complexity] = cols
    
    df = df[cols]
    df = df[sorted(df.columns)]
    df = df[[elem for elem in list(df.columns) if elem != 'happiness_ladder'] + ['happiness_ladder']]
    
    df_shuffled = df.sample(frac=1, random_state=STATE).reset_index(drop=True)
    df_shuffled.to_csv(f'data/preprocessed/korea-{complexity}.csv', index=False)
    
    num = int(len(df_shuffled) * 0.9) + 1
    df_shuffled_train = df_shuffled[:num]
    df_shuffled_test = df_shuffled[num:]

    df_shuffled_train.to_csv(f'data/preprocessed/korea-{complexity}-train.csv', index=False)
    df_shuffled_test.to_csv(f'data/preprocessed/korea-{complexity}-test.csv', index=False)
```

```python
def present(elem, complexity):
    if elem in cols_dict[complexity]:
        return 1
    else:
        return 0

code['all'] = [present(elem, 'all') for elem in code['code']]
code['complex'] = [present(elem, 'complex') for elem in code['code']]
code['medium'] = [present(elem, 'medium') for elem in code['code']]
code['basic'] = [present(elem, 'basic') for elem in code['code']]
code['simple'] = [present(elem, 'simple') for elem in code['code']]
```

```python
code.to_excel('code-stats.xlsx', index=False)
```

```python
string = "["
for _, row in code.iterrows():
    code, cat, main, sup, _, _, alls, complexs, medium, basic, simple = row
    elem = "'"
    if alls == 0:
        elem += "(N/A) "
    elif complexs == 0:
        elem += "(ALL only) "
    elif medium == 0:
        elem += "(COMPLEX+ only) "
    elif basic == 0:
        elem += "(MEDIUM+ only) "
    elif simple == 0:
        elem += "(BASIC+ only) "
    if cat == 1:
        elem += "(Categorical) "
    elem += f"{code} - {main} - {sup}', "
    elem = elem.replace(" - nan", "")
    string += elem
    
string = string[:-2]
string += "]"
print(string)
```
